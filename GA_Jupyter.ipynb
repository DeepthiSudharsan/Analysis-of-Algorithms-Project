```scala
  val n = 5;  //number of cities
  val A = Array.ofDim[Int](n, n) // adjaceny matrix
/*
  A(0)(0) = 0
A(0)(1) = 5
A(0)(2) = 8
A(0)(3) = 4
A(0)(4) = 5
A(0)(5)=7
A(0)(6)=6
A(0)(7)=4



A(1)(0) = 5
A(1)(1) = 0
A(1)(2) = 7
A(1)(3) = 4
A(1)(4) = 5
A(1)(5)=4
A(1)(6)=12
A(1)(7)=2



A(2)(0) = 8
A(2)(1) = 7
A(2)(2) = 0
A(2)(3) =8
A(2)(4) = 6
A(2)(5)=20
A(2)(6)=8
A(2)(7)=3



A(3)(0) = 4
A(3)(1) = 4
A(3)(2) =8
A(3)(3) =0
A(3)(4) = 8
A(3)(5)=2
A(3)(6)=1
A(3)(7)=3



A(4)(0) = 5
A(4)(1) = 5
A(4)(2) =6
A(4)(3) =8
A(4)(4) = 0
A(4)(5)=23
A(4)(6)=12
A(4)(7)=14



A(5)(0)=15
A(5)(1)=19
A(5)(2)=25
A(5)(3)=37
A(5)(4)=28
A(5)(5)=0
A(5)(6)=3
A(5)(7)=5



A(6)(0)=12
A(6)(1)=4
A(6)(2)=12
A(6)(3)=21
A(6)(4)=32
A(6)(5)=21
A(6)(6)=0
A(6)(7)=28



A(7)(0)=6
A(7)(1)=7
A(7)(2)=8
A(7)(3)=4
A(7)(4)=21
A(7)(5)=12
A(7)(6)=14
A(7)(7)=0
 
*/
A(0)(0) = 0
A(0)(1) = 5
A(0)(2) = 8
A(0)(3) = 4
A(0)(4) = 5

A(1)(0) = 5
A(1)(1) = 0
A(1)(2) = 7
A(1)(3) = 4
A(1)(4) = 5

A(2)(0) = 8
A(2)(1) = 7
A(2)(2) = 0
A(2)(3) =8
A(2)(4) = 6

A(3)(0) = 4
A(3)(1) = 4
A(3)(2) =8
A(3)(3) =0
A(3)(4) = 8

A(4)(0) = 5
A(4)(1) = 5
A(4)(2) =6
A(4)(3) =8
A(4)(4) = 0

  var bestFit = 9999.0
  //Step - 0 - Representation
  def fitness(city:List[Int]):Float = {
    
    def dist:Float = {
    var dist = 0
    for (i <-0 to city.length-2){
       // println(A(city(i)-1)(city(i+1)-1))
       dist = dist + A(city(i)-1)(city(i+1)-1)
    }
    //println(A(city(city.length-1)-1)(city(0)-1))
    dist = dist +A(city(city.length-1)-1)(city(0)-1)
        dist
    }
    var fitness = dist
    if(fitness < bestFit){
    bestFit = fitness
    }
    fitness
}
  
  val cities = (1 to n)
  
  //Step 1 - Generating the first random population
    var p = (0 to 100).map{ i => val city:List[Int] = (Random.shuffle(cities)).toList
    (city,fitness(city)) }
  
   var avgFit = p.map(_._2).reduce(_ + _) / p.length.toFloat
  

  //Step 2 - Fitness
  
  
   def rt = scala.util.Random.nextInt(n)
   def die = scala.util.Random.nextInt(4)
   
   def breed(p1: List[Int], p2:  List[Int]) = {
    val c1Point = rt

    // Crossover...
    var child1 = p1.splitAt(c1Point)._1 ++ p2.filter(h => !(p1.splitAt(c1Point)._1.contains(h)))
    var child2 = p2.splitAt(c1Point)._1 ++ p1.filter(h => !(p2.splitAt(c1Point)._1.contains(h)))

    // Mutation... probabilistic
    val c2Point = rt
    die match {
      case 1 => (child1.updated(c1Point,child1(c2Point)).updated(c2Point,child1(c1Point)), child2)
      case 2 => (child1,child2.updated(c1Point,child2(c2Point)).updated(c2Point,child2(c1Point)) )
      case _ => (child1, child2)
    }
 
  
 
  }
  
  
  def select = {

    // Selection
      val pGood = p.sortBy(_._2).take(25) 
      //val pGood = p.sorted(Ordering.DataType.reverse) 
      

    def rg = scala.util.Random.nextInt(pGood.length)

    // New p
    p = (0 to 500).flatMap { j =>
      val p1 = pGood(rg)
      val p2 = pGood(rg)
      if( p1._2 > bestFit || p2._2 > bestFit){
      val (c1, c2) = breed(p1._1, p2._1)
      List((c1,fitness(c1)), (c2,fitness(c2)))
      }
      else{
      val (c1, c2) = breed(pGood(rg)._1, pGood(rg)._1)
      List((c1,fitness(c1)), (c2,fitness(c2)))
      }
    }

    avgFit = p.map(_._2).reduce(_ + _) / p.length.toFloat
    avgFit

  }
  
  var i = 100;

  while (i > 0) {
    i -= 1
    println(i + " " + select )
  }
 

```
