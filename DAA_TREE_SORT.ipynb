```scala
import java.io._ 
import scala.collection.mutable.ArrayBuffer
```

```scala
case class Node[T <% Ordered[T]](var data: T, var left: Node[T], var right: Node[T]){
   override def toString = data.toString 
}
```

```scala

case class BinarySearchTree[T <% Ordered[T]](var root: Node[T],var arr: ArrayBuffer[T]) {
    
    def insert(data: T){
     if (root == null) { root = Node[T](data, null, null); return }

     var node = root      
     while(node != null){
        val res = data.compare(node.data)
        if (res == 0) return

        if (res < 0)
           if (node.left == null) { node.left = Node[T](data, null, null); return }
           else node = node.left 
        else
           if (node.right == null) { node.right = Node[T](data, null, null); return }
           else node = node.right 
     }     
  }  
    
    
  def inOrderApplyRec(apply: (Node[T]) => Unit ):ArrayBuffer[T] ={ 
      
      inOrderApply(root, apply) 
  }
    
    
  private def inOrderApply(node: Node[T], apply: (Node[T]) => Unit):ArrayBuffer[T]= {
      if (node.left!=null) inOrderApply(node.left, apply)
      apply(node)
      arr += node.data
      //print(arr)
      if (node.right!=null) inOrderApply(node.right, apply) 
      arr
  } 

        
}
```

```scala
def buildtree(A:Array[Int]):BinarySearchTree[Int]={
    var tree = BinarySearchTree[Int](null,ArrayBuffer[Int]() )
    for(i <- 0 to A.length-1){
        tree.insert(A(i));
    }
    tree
}
```
